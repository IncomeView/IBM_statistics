<a id="preparacao"></a>
## 1. Preparação de Dados (Data Preparation/Data Cleaning))

- **O que é:**  
    - Processo de limpar, padronizar e organizar os dados para modelagem.

- **Inclui:**  
    - Tratamento de valores ausentes
    - Padronização de formatos (datas, categorias, strings)
    - Remoção de duplicidades
    - Correção de inconsistências
    - Normalização / padronização
    - Junção de múltiplas fontes (merge/join)
    - Detecção e tratamento de outliers  

- **Ferramentas que uso:**  
    - **Pandas** (merge, groupby, fillna, astype, datetime)  
    - **NumPy** (operações vetorizadas, máscaras, transformações)  

- **Como eu explico na entrevista:**  
    > "Preparação de dados é garantir que a base esteja limpa e consistente. Tenho experiência prática com tratamento de nulos, padronização, integração de fontes e organização de dados para análises financeiras, risco e inadimplência."

- **Exemplo do meu histórico:**  
    - Consolidação de dados de ERP, CRM, cobrança e vendas via APIs.  
    - Padronização de bases financeiras e operacionais para FP&A e crédito.  
    - Construção de camadas raw → refined → analytics no PostgreSQL.

<a id="feature-engineering"></a>
## 2. Feature Engineering

- **O que é:**  
    - Criação e transformação de variáveis para aumentar o poder preditivo do modelo.

- **Inclui:**  
    - Criação de indicadores derivados
    - Transformações matemáticas (log, binning, normalização)
    - Encoding de variáveis categóricas
    - Criação de variáveis temporais
    - Agregações por cliente/contrato
    - Variáveis de comportamento  

- **Ferramentas que uso:**  
    - **Pandas** (rolling, shift, diff, apply)  
    - **Scikit-Learn** (OneHotEncoder, StandardScaler, PolynomialFeatures)  

- **Como eu explico na entrevista:**  
    > "Feature engineering é transformar dados brutos em informação preditiva. Minha experiência em crédito e inadimplência me ajuda a criar variáveis relevantes para risco."

- **Exemplo do meu histórico:**  
    - % de parcelas pagas em dia  
    - Aging e comportamento de carteira  
    - Indicadores de capacidade de pagamento  
    - Variáveis derivadas de fluxo de caixa e liquidez
    - enquadramentos (pagamento, inadimplencia, portadores)

<a id="validacao"></a>
## 3. Validação de Modelos (Model Validation)

- **O que é:**  
    - Avaliar se o modelo generaliza bem e não sofre overfitting.

- **Inclui:**  
    - Separação treino/teste
    - Validação cruzada (k-fold)
    - Métricas adequadas (AUC, KS, recall, precision, RMSE)
    - Comparação entre modelos
    - Avaliação de estabilidade  

- **Ferramentas que uso:**  
    - **Scikit-Learn** (train_test_split, cross_val_score, metrics)  

- **Como eu explico na entrevista:**  
    > "Validação garante que o modelo funciona fora da amostra. Uso validação cruzada e métricas adequadas para cada tipo de problema."

- **Exemplo do meu histórico:**  
    - Validação de indicadores de inadimplência  
    - Testes de estabilidade de carteira  
    - Comparação de cenários financeiros e projeções

<a id="tuning"></a>
## 4. Tuning (Hyperparameter Tuning)

- **O que é:**  
    - Ajuste fino dos hiperparâmetros para melhorar performance.

- **Inclui:**  
    - GridSearchCV
    - RandomizedSearchCV
    - Ajuste de profundidade, learning rate, número de árvores
    - Regularização (L1/L2)  

- **Ferramentas que uso:**  
    - **Scikit-Learn** (GridSearchCV, RandomizedSearchCV)  
    - **XGBoost / CatBoost** (parâmetros avançados)  

- **Como eu explico na entrevista:**  
    > "Tuning é otimizar hiperparâmetros para melhorar performance. Tenho prática com GridSearch e RandomSearch em exercícios e projetos de estudo."

- **Exemplo do meu histórico:**  
    - Ajuste de parâmetros em modelos analíticos (não ML)  
    - Otimização de KPIs financeiros e operacionais  
    - Ajuste de regras de risco e elegibilidade
