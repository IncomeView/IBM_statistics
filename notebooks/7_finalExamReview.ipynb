A seguir, exemplos diretos do projeto final.

## **5.1 T-test: Casas adjacentes ao Charles River têm MEDV diferente?**

*H0:* não há diferença  
*H1:* há diferença  

```python
scipy.stats.ttest_ind(df[df.CHAS==1]["MEDV"],
                      df[df.CHAS==0]["MEDV"],
                      equal_var=True)
```

**Interpretação esperada na prova:**  
- Se p-value < 0.05 → rejeita H0  
- Conclusão deve ser escrita em linguagem de negócio  

## **5.2 ANOVA: MEDV difere por faixas de AGE?**

```python
df["AGE_GROUP"] = pd.cut(df["AGE"],
                         bins=[0, 40, 80, 100],
                         labels=["≤40", "40–80", ">80"])

scipy.stats.f_oneway(df[df.AGE_GROUP=="≤40"]["MEDV"],
                     df[df.AGE_GROUP=="40–80"]["MEDV"],
                     df[df.AGE_GROUP==">80"]["MEDV"])
```

## **5.3 Correlation: NOX e INDUS estão relacionados?**

```python
scipy.stats.pearsonr(df["NOX"], df["INDUS"])
```

## **5.4 Interpretação esperada**

- Pearson correlation retorna `(coeficiente, p-value)`
- coeficiente > 0 → relação positiva
- coeficiente < 0 → relação negativa
- |coef| > 0.7 → forte
- p-value < 0.05 → estatisticamente significativo

## **6.1 Simple Regression: DIS → MEDV**

Pergunta do projeto:
*Qual o impacto de uma distância adicional ponderada para centros de emprego sobre o valor médio das casas?*

```python
X = sm.add_constant(df["DIS"])
y = df["MEDV"]

model = sm.OLS(y, X).fit()
model.summary()
```

## **Como interpretar na prova**

- β₁ negativo → maior distância reduz MEDV  
- p-value < 0.05 → efeito significativo  
- R² → porcentagem explicada  

## **6.2 Multiple Regression — Modelo completo**

```python
X = df.drop(columns=["MEDV"])
X = sm.add_constant(X)
y = df["MEDV"]

model_full = sm.OLS(y, X).fit()
model_full.summary()
```

## **O que você deve saber interpretar**

- **coeficiente**: impacto marginal  
- **p-value**: significância  
- **R²**: qualidade do ajuste  
- **multicolinearidade** (não cai pesado, mas pode aparecer)
