# 2. Intuição das Redes Neurais Convolucionais (CNNs)

CNNs são redes neurais projetadas para lidar com **imagens**.

Elas funcionam de forma diferente das redes densas (MLPs):

- MLPs tratam cada pixel como uma entrada independente  
- CNNs exploram **padrões espaciais** (bordas, texturas, formas)  

Isso faz CNNs serem:
- mais eficientes  
- mais rápidas  
- mais precisas  
- mais robustas  

CNNs são o padrão-ouro em visão computacional.

---
# 2.1 O problema das redes densas com imagens

Uma imagem 100x100 tem:

```
100 * 100 * 3 = 30.000 entradas
```

Uma camada densa com 128 neurônios teria:

```
30.000 * 128 = 3.840.000 pesos
```

Isso é:
- lento  
- difícil de treinar  
- propenso a overfitting  

CNNs resolvem isso com **convoluções**.

---
# 2.2 O que é uma convolução?

Uma convolução é uma operação que:

- aplica um **filtro (kernel)** pequeno  
- percorre a imagem  
- detecta padrões locais  

Exemplo de filtro 3x3:

```
[ 1  0 -1 ]
[ 1  0 -1 ]   → detector de borda vertical
[ 1  0 -1 ]
```

Quando esse filtro percorre a imagem, ele “acende” onde há bordas.

---
# 2.3 Intuição visual de uma convolução

````
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Criando uma imagem simples com uma borda vertical
img = np.zeros((20, 20))
img[:, 10:] = 1  # metade direita branca

# Filtro de borda vertical
kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

# Aplicando convolução
conv = convolve2d(img, kernel, mode="same")

plt.figure(figsize=(10, 3))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap="gray")
plt.title("Imagem Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(kernel, cmap="gray")
plt.title("Filtro (Kernel)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(conv, cmap="gray")
plt.title("Resultado da Convolução")
plt.axis("off")

plt.show()
````

🟣 **Interpretação**

- O filtro detecta a borda vertical  
- A convolução gera um mapa de ativação  
- CNNs aprendem esses filtros automaticamente  

Isso é o coração das CNNs.

---
# 2.4 Por que convoluções funcionam tão bem?

CNNs exploram três ideias fundamentais:

## ✔️ 1. Localidade
Pixels próximos têm relação entre si.

## ✔️ 2. Compartilhamento de pesos
O mesmo filtro percorre toda a imagem.

Isso reduz drasticamente o número de parâmetros.

## ✔️ 3. Invariância espacial
Um gato é um gato:
- no canto  
- no centro  
- em cima  
- embaixo  

CNNs detectam padrões independentemente da posição.

---
# 2.5 CNNs aprendem filtros automaticamente

Em vez de você definir filtros (como bordas, cantos, texturas),
a rede aprende sozinha.

Camadas iniciais aprendem:
- bordas  
- linhas  
- cantos  

Camadas intermediárias aprendem:
- texturas  
- padrões locais  

Camadas profundas aprendem:
- partes de objetos  
- objetos inteiros  

---
# 2.6 Exemplo real: filtros aprendidos por CNNs

Vamos visualizar filtros de uma CNN simples.

````
# Criando uma CNN mínima
cnn = keras.Sequential([
    layers.Conv2D(8, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(16, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(10, activation="softmax")
])

cnn.summary()
````

🟣 **Interpretação**

- A primeira camada aprende 8 filtros 3x3  
- A segunda camada aprende 16 filtros 3x3  
- Cada filtro detecta um padrão diferente  

CNNs constroem uma hierarquia de features.

---

✔️ Filtros percorrem a imagem e criam mapas de ativação  
✔️ CNNs aprendem filtros automaticamente  
✔️ São muito mais eficientes que redes densas  
✔️ São o padrão-ouro em visão computacional

# 3. Pooling e Extração de Features

Depois da convolução, CNNs aplicam **pooling**.

Pooling reduz a resolução espacial da imagem, mantendo apenas as
informações mais importantes.

Isso traz:
- menos parâmetros  
- menos risco de overfitting  
- invariância a pequenas mudanças na imagem  
- maior velocidade  

---
# 3.1 Max Pooling — o mais usado

Max Pooling pega o **maior valor** dentro de uma janela.

Exemplo com janela 2×2:

```
[1 3]
[2 0]   → max = 3
```

Ele mantém apenas a ativação mais forte (o padrão mais evidente).

````
import numpy as np
import matplotlib.pyplot as plt

# Criando um mapa de ativação simples
activation = np.array([
    [0, 1, 2, 1],
    [2, 3, 1, 0],
    [1, 0, 2, 4],
    [0, 1, 3, 2]
])

# Max pooling manual 2x2
pooled = np.array([
    [np.max(activation[0:2, 0:2]), np.max(activation[0:2, 2:4])],
    [np.max(activation[2:4, 0:2]), np.max(activation[2:4, 2:4])]
])

plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.imshow(activation, cmap="viridis")
plt.title("Mapa de Ativação (4×4)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(pooled, cmap="viridis")
plt.title("Após Max Pooling (2×2)")
plt.axis("off")

plt.show()
````

🟣 **Interpretação**

- O pooling reduz a imagem pela metade  
- Mantém apenas os padrões mais fortes  
- Remove ruído e detalhes irrelevantes  

---
# 3.2 Average Pooling

Em vez do máximo, ele tira a **média** da janela.

É menos usado em CNNs modernas, mas útil em:
- redes antigas (LeNet)  
- algumas arquiteturas de segmentação  
- Global Average Pooling (muito importante!)  

````
avg_pooled = np.array([
    [np.mean(activation[0:2, 0:2]), np.mean(activation[0:2, 2:4])],
    [np.mean(activation[2:4, 0:2]), np.mean(activation[2:4, 2:4])]
])

plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.imshow(activation, cmap="magma")
plt.title("Mapa de Ativação (4×4)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(avg_pooled, cmap="magma")
plt.title("Após Average Pooling (2×2)")
plt.axis("off")

plt.show()
````

---
# 3.3 Global Average Pooling (GAP)

GAP reduz **cada mapa de ativação** para **um único número**.

Se a camada anterior tem 128 filtros:
- GAP → vira um vetor de 128 números  

Isso substitui camadas densas enormes.

É usado em:
- ResNet  
- MobileNet  
- EfficientNet  
- Vision Transformers (na etapa final)  

````
# Exemplo simples de GAP
feature_map = np.random.rand(8, 8)  # mapa 8x8
gap_value = np.mean(feature_map)

gap_value
````

🟣 **Por que GAP é tão importante?**

✔️ reduz drasticamente o número de parâmetros  
✔️ evita overfitting  
✔️ funciona muito bem em redes profundas  
✔️ substitui camadas densas gigantes  

---
# 3.4 Pooling traz invariância espacial

Se um objeto se move um pouco na imagem:
- a convolução detecta o padrão  
- o pooling garante que a posição exata não importa  

Isso é essencial para visão computacional.

---
# 3.5 Pooling na prática com Keras

````
from tensorflow.keras import layers, models

cnn_pool = models.Sequential([
    layers.Conv2D(8, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(16, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(10, activation="softmax")
])

cnn_pool.summary()
````

🟣 **Interpretação**

- Cada MaxPooling reduz a imagem pela metade  
- A rede fica mais rápida e mais leve  
- A extração de features fica mais robusta  

---
# 3.6 Conclusão desta parte

✔️ Pooling reduz dimensionalidade  
✔️ MaxPooling é o mais usado  
✔️ Average Pooling ainda é útil em alguns casos  
✔️ Global Average Pooling é essencial em redes modernas  
✔️ Pooling traz invariância espacial

# 4. Construindo sua primeira CNN completa

Agora vamos treinar uma CNN real usando o dataset MNIST.

A arquitetura será:

- Conv2D(32 filtros, kernel 3×3, ReLU)
- MaxPooling2D(2×2)
- Conv2D(64 filtros, kernel 3×3, ReLU)
- MaxPooling2D(2×2)
- Flatten
- Dense(64, ReLU)
- Dense(10, Softmax)

Essa é uma CNN clássica para classificação de imagens.

---
# 4.1 Carregando o dataset MNIST

````
from tensorflow.keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train.shape, y_train.shape
````

---
# 4.2 Visualizando algumas imagens

````
plt.figure(figsize=(8, 3))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i], cmap="gray")
    plt.axis("off")
plt.suptitle("Exemplos do dataset MNIST")
plt.show()
````

---
# 4.3 Pré-processamento

CNNs esperam:
- imagens normalizadas (0–1)
- formato (altura, largura, canais)

````
# Normalizando
X_train = X_train / 255.0
X_test = X_test / 255.0

# Expandindo dimensão para canal único
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)

X_train.shape, X_test.shape
````

---
# 4.4 Construindo a CNN

````
model_cnn = keras.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])

model_cnn.summary()
````

---
# 4.5 Compilando o modelo

````
model_cnn.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
````

---
# 4.6 Treinando a CNN

````
history_cnn = model_cnn.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.1,
    verbose=1
)
````

---
# 4.7 Curvas de aprendizado

````
plt.figure(figsize=(8, 5))
plt.plot(history_cnn.history["accuracy"], label="Treino")
plt.plot(history_cnn.history["val_accuracy"], label="Validação")
plt.title("Acurácia — CNN")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history_cnn.history["loss"], label="Treino")
plt.plot(history_cnn.history["val_loss"], label="Validação")
plt.title("Perda — CNN")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.8 Avaliação no conjunto de teste

````
test_loss, test_acc = model_cnn.evaluate(X_test, y_test, verbose=0)
test_acc
````

🟣 **Interpretação**

CNNs atingem facilmente **98–99% de acurácia** no MNIST.

Isso é muito superior a:
- regressão logística  
- MLP simples  
- SVM sem kernel  

CNNs são incrivelmente eficientes para imagens.

---
# 4.9 Fazendo previsões

````
preds = model_cnn.predict(X_test[:10])
pred_labels = np.argmax(preds, axis=1)

plt.figure(figsize=(10, 3))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap="gray")
    plt.title(f"Pred: {pred_labels[i]}")
    plt.axis("off")
plt.show()
````

---
# 4.10 Conclusão desta parte

✔️ Construímos nossa primeira CNN real  
✔️ Treinamos no MNIST  
✔️ Obtivemos alta acurácia  
✔️ Visualizamos previsões  
✔️ Entendemos o fluxo completo de uma CNN

# 5. Data Augmentation e Treinamento Avançado

Data Augmentation é uma técnica essencial em visão computacional.

Ela cria **novas imagens artificiais** a partir das imagens reais,
aplicando transformações como:

- rotações  
- zoom  
- deslocamentos  
- inversões  
- brilho  
- ruído  

Isso aumenta a robustez da CNN e reduz overfitting.

---
# 5.1 Por que Data Augmentation é tão importante?

CNNs são muito poderosas e podem memorizar facilmente o dataset.

Data Augmentation:

✔️ aumenta o dataset  
✔️ melhora generalização  
✔️ reduz overfitting  
✔️ simula variações do mundo real  
✔️ torna o modelo mais robusto  

É obrigatório em praticamente todos os projetos de visão computacional.

---
# 5.2 Criando um gerador de Data Augmentation com Keras

Vamos usar `ImageDataGenerator`, que aplica transformações em tempo real.

````
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

# Ajustando ao dataset
datagen.fit(X_train)
````

---
# 5.3 Visualizando Data Augmentation

````
# Pegando uma imagem
img = X_train[0]
img = np.expand_dims(img, 0)

# Gerando 10 versões aumentadas
plt.figure(figsize=(10, 3))
for i, batch in enumerate(datagen.flow(img, batch_size=1)):
    plt.subplot(2, 5, i+1)
    plt.imshow(batch[0].reshape(28,28), cmap="gray")
    plt.axis("off")
    if i == 9:
        break
plt.suptitle("Exemplos de Data Augmentation")
plt.show()
````

🟣 **Interpretação**

A imagem original é transformada de várias maneiras:
- levemente rotacionada  
- deslocada  
- ampliada  

Isso ajuda a CNN a aprender padrões mais robustos.

---
# 5.4 Construindo uma CNN com Data Augmentation

````
model_aug = keras.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])

model_aug.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
````

---
# 5.5 Treinando com Data Augmentation

````
history_aug = model_aug.fit(
    datagen.flow(X_train, y_train, batch_size=64),
    epochs=8,
    validation_data=(X_test, y_test),
    verbose=1
)
````

---
# 5.6 Curvas de aprendizado com Data Augmentation

````
plt.figure(figsize=(8, 5))
plt.plot(history_aug.history["accuracy"], label="Treino")
plt.plot(history_aug.history["val_accuracy"], label="Validação")
plt.title("Acurácia — CNN com Data Augmentation")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history_aug.history["loss"], label="Treino")
plt.plot(history_aug.history["val_loss"], label="Validação")
plt.title("Perda — CNN com Data Augmentation")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

🟣 **Interpretação**

- Validação melhora mais lentamente  
- Treino e validação ficam mais próximos  
- Overfitting reduz drasticamente  

Isso é exatamente o que queremos.

---
# 5.7 Avaliação no conjunto de teste

````
test_loss_aug, test_acc_aug = model_aug.evaluate(X_test, y_test, verbose=0)
test_acc_aug
````

🟣 **Resultado típico**

- CNN sem augmentation → 98%  
- CNN com augmentation → 98% também, mas **muito mais robusta**  

Em datasets reais, augmentation pode melhorar 5–20%.

---

✔️ Aprendemos Data Augmentation  
✔️ Visualizamos transformações reais  
✔️ Treinamos uma CNN com augmentation

# 7. Transfer Learning com modelos pré‑treinados

Transfer Learning significa:

> Usar um modelo treinado em milhões de imagens (ImageNet)
> e adaptá‑lo para o seu problema.

Isso funciona porque:
- as primeiras camadas aprendem bordas, texturas, formas  
- as camadas intermediárias aprendem partes de objetos  
- as camadas finais aprendem conceitos mais abstratos  

Você só precisa treinar a parte final.

---
# 7.1 Escolhendo um modelo pré‑treinado

Vamos usar **MobileNetV2**, um modelo rápido e eficiente.

Ele foi treinado no ImageNet (1.2 milhão de imagens, 1000 classes).

# 8. Projeto Final — Classificador de Imagens Completo

Neste projeto, vamos integrar:

✔️ Transfer Learning

Usaremos o dataset **MNIST**, mas com pipeline profissional.

---
# 8.1 Preparação dos dados

````
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalizando
X_train = X_train / 255.0
X_test = X_test / 255.0

# Expandindo canal
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
````

---
# 8.2 Data Augmentation profissional

````
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.15
)

datagen.fit(X_train)
````

---
# 8.3 CNN do zero (baseline)

````
cnn_baseline = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])

cnn_baseline.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_base = cnn_baseline.fit(
    datagen.flow(X_train, y_train, batch_size=64),
    epochs=6,
    validation_data=(X_test, y_test),
    verbose=1
)
````

---
# 8.4 Avaliação do baseline

````
loss_base, acc_base = cnn_baseline.evaluate(X_test, y_test, verbose=0)
acc_base
````

---
# 8.5 Transfer Learning com MobileNetV2

````
from tensorflow.keras.applications import MobileNetV2

base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(96, 96, 3)
)

base_model.trainable = False

model_tl = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])

model_tl.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
````

---
# 8.6 Preparando MNIST para Transfer Learning

````
# Redimensionando e convertendo para RGB
X_train_resized = tf.image.resize(X_train, (96, 96))
X_test_resized = tf.image.resize(X_test, (96, 96))

X_train_rgb = tf.image.grayscale_to_rgb(X_train_resized)
X_test_rgb = tf.image.grayscale_to_rgb(X_test_resized)

# Preprocessamento MobileNetV2
X_train_pre = preprocess_input(X_train_rgb)
X_test_pre = preprocess_input(X_test_rgb)
````

---
# 8.7 Treinando Transfer Learning

````
history_tl = model_tl.fit(
    X_train_pre, y_train,
    validation_split=0.1,
    epochs=5,
    batch_size=64,
    verbose=1
)
````

---
# 8.8 Avaliação do Transfer Learning

````
loss_tl, acc_tl = model_tl.evaluate(X_test_pre, y_test, verbose=0)
acc_tl
````

---
# 8.9 Fine‑tuning (descongelando camadas profundas)

````
base_model.trainable = True

for layer in base_model.layers[:100]:
    layer.trainable = False

model_tl.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_ft = model_tl.fit(
    X_train_pre, y_train,
    validation_split=0.1,
    epochs=3,
    batch_size=64,
    verbose=1
)
````

---
# 8.10 Avaliação final após Fine‑tuning

````
loss_ft, acc_ft = model_tl.evaluate(X_test_pre, y_test, verbose=0)
acc_ft
````

---
# 8.11 Comparação entre modelos
