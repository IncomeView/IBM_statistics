# 1. Introdução à Regressão

🔵 **Conteúdo oficial do módulo 6 (IBM)**

O curso começa afirmando:

> - “A regressão é o carro‑chefe da análise estatística.”

E explica que, enquanto testes de hipótese analisam relações **duas a duas**,  
a regressão permite analisar **várias relações simultaneamente**.

Além disso:

- regressão substitui **teste t**,  
- regressão substitui **ANOVA**,  
- regressão substitui **correlação de Pearson**,  
- e ainda permite **previsão**, algo que os testes não fazem.

Vamos construir essa base.

## 1.1 O que é regressão?

Regressão é um método estatístico para modelar a relação entre:

- uma **variável dependente** (Y)  
- uma ou mais **variáveis explicativas** (X)

Exemplos do curso IBM:

- A avaliação de ensino (`eval`) pode ser explicada por:
  - beleza (`beauty`)
  - gênero (`gender`)
  - proficiência em inglês (`native`)

| Formalmente: | $ \to \qquad Y = f(X) $ |
|-|-|

## 1.2 Variável dependente (Y)

É a variável que queremos explicar ou prever.

Exemplos:
- `eval` (nota de avaliação)
- `beauty` (pontuação de beleza)
- `age` (idade)

| No módulo 6, o foco é: | $ \to \qquad Y = \text{eval} $ |
|-|-|

## 1.3 Variáveis explicativas (X)

São as variáveis que explicam a variação de Y.

Exemplos:
- `beauty`
- `gender`
- `native`
- `age`
- `students`

| O curso IBM usa principalmente: | $ \to \qquad X = \text{beauty} $ |
|-|-|

## 1.4 Modelo de regressão linear simples

| O modelo mais básico é: | $ \to \qquad Y = \beta_0 + \beta_1 X + \varepsilon $ |
|-|-|

Onde:

- $\beta_0$ = intercepto  
- $\beta_1$ = inclinação/Peso das variáveis (efeito de X sobre Y)  
- $\varepsilon$ = termo de erro (tudo que o modelo não explica | Diferença entre valor real e previsto)

O curso IBM enfatiza:

> - “O termo de erro captura tudo o que o modelo não conseguiu explicar.”

## 1.5 Modelo de regressão múltipla

| Regressão múltipla | Modelo |
|-|-|
| Quando há mais de uma variável explicativa: | $ \to \qquad Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \varepsilon $ |
| Exemplo: | $ \to \qquad \text{eval} = \beta_0 + \beta_1 \text{beauty} + \beta_2 \text{gender} + \beta_3 \text{native} + \varepsilon $ |


## 1.6 Exemplo IBM — Regressão de `eval` sobre `beauty`

### 🟠 Explicação avançada — O que é o termo de erro?

O termo de erro $\varepsilon$ representa:

- variáveis não incluídas no modelo,  
- ruído aleatório,  
- fatores não observados,  
- erros de medição.

| Formalmente: | $ \to \qquad \varepsilon_i = Y_i - \hat{Y}_i $ |
|-|-|

- Onde $\hat{Y}_i$ é o valor previsto pelo modelo.

Em regressão, analisamos os resíduos para verificar:
- normalidade,  
- homocedasticidade,  
- independência.



## 🎓 1.8 – Como o modelo de regressão foi criado <br> $\qquad$  e por que ele marca o início da Estatística Moderna


A regressão linear tem origem no final do século XIX, com Francis Galton e Karl Pearson.  
Galton observou que filhos de pais muito altos tendiam a ser menos altos, e filhos de pais muito baixos tendiam a ser menos baixos. Ele chamou esse fenômeno de:

$\to\qquad \text{regression toward the mean}$

- Pearson então formalizou matematicamente essa ideia introduzindo:
  - correlação,
  - covariância,
  - e o método dos mínimos quadrados.

|A partir disso, o modelo de regressão passou a ser escrito como:|$\to\qquad y = \beta_0 + \beta_1 x + \varepsilon$|
|-|-|

- onde o objetivo é encontrar os coeficientes $\beta_0$ e $\beta_1$ que melhor explicam a relação entre $x$ e $y$.

<br>

 1. **<u> O modelo é criado pela minimização dos erros </u>**

|Mínimo|Quadrado|
|-|-|
|O método dos Mínimos Quadrados Ordinários (OLS) busca minimizar:|$\to\qquad S(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$|
|Em forma matricial:|$\to\qquad S(\beta) = (y - X\beta)'(y - X\beta)$|
|Derivando e igualando a zero:|$\to\qquad \frac{\partial S}{\partial \beta} = -2X'(y - X\beta) = 0$|
|o que leva às **equações normais**:|$\to\qquad X'X\beta = X'y$|
|E a solução fechada é:|$\to\qquad \beta = (X'X)^{-1}X'y$|

- Esse é o estimador OLS.

<br>



2. **<u> Por que isso é considerado o início da Estatística Moderna? </u>**

A regressão introduziu três ideias fundamentais:

- <u> a. Modelagem matemática de fenômenos reais </u>  
Antes disso, estatística era apenas contagem e tabelas.  

|A regressão trouxe:|$\to\qquad \text{dados} = \text{modelo} + \text{erro}$|
|-|-|

Essa estrutura é a base de toda a estatística moderna.

- <u> b. Inferência estatística automática </u>  

|Estatística|Automática|
|-|-|
|Ao estimar $\beta_1$, a regressão automaticamente testa:|$\to\qquad H_0: \beta_1 = 0$|
|usando a estatística t:|$$\to\qquad t = \frac{\hat{\beta}_1}{SE(\hat{\beta}_1)}$$|

Ou seja, o modelo já vem com o teste embutido.

- <u> c. Conexão profunda com correlação </u>  

|Em regressão simples, o teste t é matematicamente idêntico ao teste de Pearson:|$$\to\qquad t = r \sqrt{\frac{n-2}{1-r^2}}$$|
|-|-|

Assim, regressão simples e correlação testam a mesma hipótese.

<br>

 **<u> Resumo </u>**

- A regressão nasceu com Galton e foi formalizada por Pearson.  
- O modelo é obtido minimizando a soma dos erros ao quadrado.  

|A solução é dada por:|$\to\qquad \beta = (X'X)^{-1}X'y$|
|-|-|


- A regressão testa automaticamente hipóteses sobre os coeficientes.  
- Ela marca o início da estatística moderna porque introduziu:
  - modelagem matemática,
  - inferência estatística,
  - estrutura “modelo + erro”,
  - e a conexão entre variáveis via correlação e regressão.

#### 🟠 Explicação avançada — Por que verificar a normalidade dos resíduos?

Quando ajustamos um modelo de regressão linear, não basta olhar apenas para os coeficientes ou para o R².  
A etapa crítica é avaliar os **resíduos**, pois são eles que determinam se a inferência estatística do modelo é confiável.

- **<u> O ponto central </u>**  
A normalidade dos resíduos **não valida o modelo em si**.  
O que ela valida são os **testes estatísticos usados dentro da regressão**, como:
    - testes t para os coeficientes  
    - teste F para o modelo  
    - p‑values  
    - intervalos de confiança  

|Esses testes assumem que o termo de erro segue:|$ \to \qquad \varepsilon \sim N(0, \sigma^2) $|
|-|-|

- Se essa suposição é razoável, então:  
    - os coeficientes estimados têm distribuição aproximadamente normal  
    - os p‑values são confiáveis  
    - os testes t e F funcionam corretamente  
    - os intervalos de confiança são válidos  

<br>

- **<u> Por que isso importa? </u>**  
Porque a regressão linear clássica usa **inferência paramétrica**.  
Se os resíduos forem muito distantes da normalidade (caudas pesadas, assimetria forte, bimodalidade), os testes podem:
    - subestimar ou superestimar significância  
    - gerar p‑values incorretos  
    - produzir intervalos de confiança distorcidos  

<br>

- **<u> E quando os resíduos são aproximadamente normais? </u>**  
Quando o histograma mostra uma forma próxima da normal:
    - os erros parecem aleatórios  
    - não há padrões estruturados  
    - a variância é estável  
    - a inferência estatística pode ser levada a sério  

Ou seja:

👉 **A normalidade dos resíduos garante que as conclusões do modelo são estatisticamente confiáveis.**


- **<u> Resumo </u>**  
    - Normalidade dos resíduos **não valida o modelo**, valida a **inferência**.  
    - Se os resíduos forem aproximadamente normais, os testes t e F são confiáveis.  
    - Isso significa que os p‑values e intervalos de confiança podem ser interpretados com segurança.




## 2.6 Regressão como estimador de efeito

O coeficiente β₁ é interpretado como:

> - “O efeito médio de X sobre Y.”

|No exemplo:|$ \to \qquad \beta_1 = 0.133 $|
|-|-|

Isso significa:

- cada aumento de 1 ponto em beleza → +0.133 em eval  
- efeito pequeno, mas estatisticamente significativo  

#### 🟠 Explicação avançada — Como é desenvolvido o cálculo de β (beta) na regressão linear

O objetivo da regressão linear é encontrar os coeficientes $\beta$ que melhor explicam a relação entre as variáveis explicativas $(X)$ e a variável resposta $(y)$.  
Esses coeficientes são obtidos pelo método dos **Mínimos Quadrados Ordinários (OLS - ordinary least squares)**.

A ideia central é simples:

👉 Encontrar os valores de $\beta$ que **minimizam a soma dos erros ao quadrado**.





1. **<u> O modelo em forma matricial </u>** 

|A regressão linear pode ser escrita como:|$ \to \qquad y = X\beta + \varepsilon$|
|-|-|

- onde:
    - $y \to$ é o vetor de valores observados  
    - $X \to$ é a matriz de preditores (inclui uma coluna de 1 para o intercepto)  
    - $\beta \to$ é o vetor de coeficientes desconhecidos  
    - $\varepsilon \to$ é o vetor de erros  

<br>

2. **<u> A função que queremos minimizar </u>**

|O erro total do modelo é a soma dos quadrados dos resíduos:|$\to \qquad S(\beta) = (y - X\beta)'(y - X\beta)$|
|-|-|

- O objetivo é encontrar o $\beta$ que minimiza $S(\beta)$.

<br>



3. **<u> Derivando para encontrar o mínimo </u>**

|Mínimo|Derivada|
|-|-|
| Para minimizar $S(\beta)$, derivamos em relação a $\beta$:| $$\to \qquad \frac{\partial S}{\partial \beta} = -2X'(y - X\beta)$$ |
| Igualamos a zero (condição de mínimo):| $\to \qquad X'(y - X\beta) = 0$ |
| Reorganizando: | $\to \qquad X'X\beta = X'y$ |

- Essas são as **equações normais** da regressão linear.

<br>

4. **<u> Solução fechada (fórmula do $\beta$) </u>**

|Se a matriz $X'X$ é invertível, a solução é:|$\to \qquad \beta = (X'X)^{-1}X'y$ |
|-|-|

Essa é a fórmula clássica dos Mínimos Quadrados Ordinários (OLS).

<br>



5. **<u> Intuição do cálculo </u>**

- $X'y$ mede a **correlação entre cada variável explicativa e o $y$**.  
- $X'X$ mede a **correlação entre as próprias variáveis explicativas**.  
- A inversão de $X'X$ corrige redundâncias e colinearidade.  
- O produto final gera os coeficientes que **minimizam o erro quadrático total**.

Em termos intuitivos:

👉 $\beta$ é o conjunto de coeficientes que melhor ajusta a linha (ou hiperplano) aos dados, no sentido de minimizar a soma dos erros ao quadrado.

<br>

6. **<u> Por que essa solução é importante? </u>**

- Quando os resíduos têm variância constante e média zero, o estimador OLS:
    - é **não enviesado**  
    - tem **variância mínima** entre todos os estimadores lineares  
    - é **eficiente** quando os resíduos são normais  

Essas propriedades são garantidas pelo **Teorema de Gauss–Markov**.

<br>

**<u> Resumo </u>** 

>- O cálculo de $\beta$ vem da minimização da soma dos erros ao quadrado.  
>- A derivação leva às equações normais: $X'X\beta = X'y$.  
>- A solução fechada é $\beta = (X'X)^{-1}X'y$.  
>- Esse $\beta$ é o melhor estimador linear possível sob as hipóteses clássicas da regressão.
>


## 2.7 Regressão como teste de hipótese

| A regressão testa automaticamente: | $ \to \qquad H_0: \beta_1 = 0 $ <br> $\to \qquad H_a: \beta_1 \ne 0 $ |
|-|-|

O valor‑p do coeficiente β₁ é exatamente o mesmo valor‑p do teste de Pearson.

#### 🟠 Explicação avançada — Por que a regressão “testa a hipótese automaticamente” <br> $\qquad\qquad\qquad\qquad\qquad$ e por que isso é exatamente o teste de Pearson

Quando ajustamos uma regressão linear simples, o software não apenas estima os coeficientes  
ele também realiza **automaticamente um teste de hipótese** sobre o coeficiente da inclinação $(\beta_1)$. Esse teste é essencial para responder:

$\to \qquad$ **“Existe relação linear real entre x e y, ou é apenas ruído?”**

A seguir, explicamos por que isso acontece e por que esse teste é matematicamente equivalente ao teste de correlação de Pearson.

<br>

1. **<u> A hipótese testada automaticamente pela regressão </u>**

|Hipótese|Teste|
|-|-|
|Ao ajustar o modelo:|$\to \qquad y = \beta_0 + \beta_1 x + \varepsilon $|
|a regressão automaticamente testa:|$\to \qquad H_0: \beta_1 = 0$ <br> $\to \qquad H_1: \beta_1 \neq 0$|

- Interpretando:
  - $(H_0)$: não existe relação linear entre x e y  
  - $(H_1)$: existe relação linear entre x e y  

Ou seja, o modelo verifica se a inclinação é realmente diferente de zero.  
- Ao estimar $\beta_1$, o modelo também calcula:  
  - erro padrão
  - estatística t
  - p‑value

E isso é exatamente um teste de hipótese.

<br>



2. **<u> O teste usado é o **teste t** do coeficiente </u>**

|A regressão calcula a estatística t:|$$\to \qquad t = \frac{\hat{\beta}_1}{SE(\hat{\beta}_1)}$$ |
|-|-|

- e, a partir dela, obtém o **p‑value**.
- Esse teste responde:  
👉 “O coeficiente é estatisticamente diferente de zero?”
  - Se sim → x tem efeito real sobre y
  - Se não → x não explica y

<br>

3. **<u> Por que isso é exatamente o teste de Pearson? </u>**

> Aqui esta a parte bonita da estatística:

Na regressão linear **com apenas um preditor**, o teste t da regressão é **matematicamente idêntico** ao teste de **correlação de Pearson**.

|A relação exata é:|$$\to \qquad t = r \sqrt{\frac{n - 2}{1 - r^2}}$$|
|-|-|

- onde:
  - $(r)$ é o coeficiente de correlação de Pearson  
  - $(n)$ é o número de observações  

- E o **p‑value** obtido é o mesmo.
- Isso significa:
  - Testar $H_0: \beta_1 = 0$ é exatamente o mesmo que testar $H_0: r = 0$

- Portanto:
  - regressão simples e correlação testam a **mesma hipótese**  
    - regressão simples = correlação + direção + escala
  - o teste estatístico é o mesmo  
  - o p‑value é o mesmo  
  - a conclusão é a mesma

- A diferença é que a regressão fornece também:
  - direção da relação  
  - magnitude da inclinação  
  - interpretação em unidades reais  

<br>



4. **<u> Intuição clara </u>**

- Se não há correlação entre $x$ e $y$:
  - a regressão encontra inclinação zero  
  - o teste t não rejeita $H_0$  
  - o coeficiente não é significativo  

- Se há correlação:
  - a regressão encontra inclinação diferente de zero  
  - o teste t rejeita $H_0$  
  - o coeficiente é significativo  

- Assim:
  - 👉 **Regressão simples é correlação com roupa de gala.**


<br>

**<u> Resumo </u>**

>- A regressão testa automaticamente $H_0: \beta_1 = 0$.  
>- O teste usado é o **teste t** do coeficiente.  
>- Em regressão simples, esse teste é **idêntico** ao teste de Pearson.  
>- Portanto, regressão simples e correlação testam a mesma hipótese:  
>   - $ H_0: \text{não existe relação linear entre x e y} $

#### 🟠 Explicação avançada — Mínimos Quadrados Ordinários (OLS)

|OLS|Função|
|-|-|
|O método OLS escolhe β₀ e β₁ que minimizam:|$\to\qquad \sum_{i=1}^n (Y_i - \hat{Y}_i)^2 $|
|As soluções fechadas são:|$$\to\qquad \beta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})} {\sum (x_i - \bar{x})^2} $$  $\to\qquad \beta_0 = \bar{y} - \beta_1 \bar{x} $|

- Isso conecta regressão diretamente com correlação.

# 3. Regressão no lugar do Teste T

🔵 **Conteúdo oficial do módulo 6 (IBM)**

O curso demonstra que:

> - “Podemos usar regressão linear no lugar do teste t.”

A ideia é simples:

- O teste t compara **duas médias**.  
- A regressão com uma variável dummy também compara **duas médias**.  

E o mais importante:

> “A estatística t e o valor‑p da regressão são idênticos aos do teste t.”

Vamos reproduzir exatamente o exemplo IBM:

- **<u> Pergunta: </u>**  
>    - Existe diferença estatisticamente significativa entre as notas de avaliação de ensino de homens e mulheres?

## 3.1 Criando a variável dummy (IBM)

- O curso transforma:
    - female = 1  
    - male = 0  

Vamos fazer o mesmo.

```python
ratings_df["female_dummy"] = (ratings_df["gender"] == "female").astype(int)
ratings_df[["gender", "female_dummy"]].head()
```

## 3.2 Teste T tradicional (para comparação)

Vamos rodar o teste t para confirmar o valor‑p.

```python
from scipy.stats import ttest_ind

female = ratings_df[ratings_df["gender"] == "female"]["eval"]
male   = ratings_df[ratings_df["gender"] == "male"]["eval"]

t_stat, p_value = ttest_ind(female, male, equal_var=True)
t_stat, p_value
```

🟣 **Interpretação**

- valor‑p < 0.05  
- diferença estatisticamente significativa  

Agora vamos ver como a regressão chega ao mesmo resultado.

## 3.3 Regressão linear com variável dummy

|Modelo:|$\to \qquad \text{eval} = \beta_0 + \beta_1 \cdot \text{female} + \varepsilon$|
|-|-|

Onde:
- β₀ = média dos homens  
- β₁ = diferença entre mulheres e homens  

Vamos ajustar o modelo.

```python
model_dummy = smf.ols("eval ~ female_dummy", data=ratings_df).fit()
model_dummy.summary()
```

## 3.4 Interpretando os coeficientes

🔵 **Conteúdo IBM**

O curso afirma:

> - “O coeficiente significa que é mais provável que você perca cerca de 0,17 pontos por ser mulher.”

Vamos interpretar:

- **Intercepto (β₀ ≈ 4.06)** $\to\qquad$ média dos homens  
- **Coeficiente (β₁ ≈ −0.17)** $\to\qquad$ mulheres têm, em média, 0.17 pontos a menos  
- **Estatística t ≈ −3.25**  
- **Valor‑p < 0.05**  

Isso é exatamente o que o curso IBM mostra.

#### 🟠 Explicação avançada <br> $\qquad\to$ Por que a regressão com variável dummy produz o mesmo resultado do teste t?

> - **Uma das conexões mais bonitas da estatística**

Quando temos uma variável categórica com dois grupos (ex.: gênero, tratamento x controle), podemos comparar as médias de duas formas:

1. usando o **teste t clássico** de comparação de médias  
2. usando **regressão linear com variável dummy**

O resultado (estatística t e p‑value) é exatamente o mesmo.  
A seguir explicamos por quê.

<br>

1. **<u> Criando a variável dummy </u>**

Suponha uma variável categórica com dois grupos:

- Grupo A  
- Grupo B  


|Variável|Dummy|
|-|-|
|Criamos uma dummy:|$\to\qquad D = \begin{cases} 1, & \text{se pertence ao grupo B} \\ 0, & \text{se pertence ao grupo A} \end{cases}$|
|Rodamos a regressão:|$$\to\qquad y = \beta_0 + \beta_1 D + \varepsilon$$|



<br>




2. **<u> O que a regressão está estimando? </u>**

|Regressão|Estimando|
|-|-|
| Quando $D = 0$:|$\to\qquad E[y\|D=0] = \beta_0 $|
|Quando $D = 1$:|$\to\qquad E[y\|D=1] = \beta_0 + \beta_1$|
|Logo:|$\to\qquad \beta_1 = E[y\|D=1] - E[y\|D=0]$|

- Ou seja:
    - 👉 **o coeficiente da dummy é exatamente a diferença entre as médias dos dois grupos.**

<br>

3. **<u> O que o teste t da regressão está testando? </u>**

|Regressão|Testando|
|-|-|
|A regressão testa automaticamente:|$\to\qquad H_0: \beta_1 = 0$|
|Que é equivalente a:|$\to\qquad H_0: E[y\|D=1] = E[y\|D=0]$|

- Ou seja:
    - 👉 **a regressão está testando se as médias dos dois grupos são iguais.**  
    Isso é exatamente o que o teste t clássico testa.

<br>



4. **<u> Por que a estatística t é idêntica? </u>**

|Idêntica|Estatística|
|-|-|
|A estatística t da regressão é:|$$\to\qquad t = \frac{\hat{\beta}_1}{SE(\hat{\beta}_1)}$$|
|Mas como:|$\to\qquad \hat{\beta}_1 = \bar{y}_1 - \bar{y}_0$|
| e o erro padrão do coeficiente é:|$\to\qquad SE(\hat{\beta}_1) = \sqrt{s^2\left(\frac{1}{n_1} + \frac{1}{n_0}\right)}$|
|então:|$$\to\qquad t = \frac{\bar{y}_1 - \bar{y}_0}{\sqrt{s^2\left(\frac{1}{n_1} + \frac{1}{n_0}\right)}}$$|



- que é **exatamente** a fórmula do teste t de Student para duas amostras independentes com variâncias iguais.

Portanto:

- 👉 **o t da regressão = t do teste t clássico**  
- 👉 **o p‑value é o mesmo**  
- 👉 **a conclusão é a mesma**

<br>

5. **<u> Intuição clara </u>**
    - A dummy separa os dados em dois grupos.  
    - A regressão calcula a diferença entre as médias.  
    - O teste t da regressão verifica se essa diferença é estatisticamente diferente de zero.  
    - Isso é exatamente o que o teste t faz.

Por isso:
> - $\to\qquad$ **Regressão com dummy é apenas outra forma de rodar o teste t.**

<br>

**<u> Resumo </u>**
>- A dummy transforma um grupo categórico em variável numérica (0/1).  
>- O coeficiente da dummy é a diferença entre as médias dos grupos.  
>- O teste t da regressão testa se essa diferença é zero.  
>- Isso é matematicamente idêntico ao teste t clássico.  
>- Por isso, regressão com dummy e teste t produzem o mesmo resultado.

<br>


## 3.5 Comparação direta: Teste T x Regressão

| Método | Estatística t | Valor‑p | Conclusão |
|--------|---------------|---------|-----------|
| Teste T | mesmo valor | mesmo valor | diferença significativa |
| Regressão | mesmo valor | mesmo valor | diferença significativa |

🟣 **Conclusão:**  
- A regressão reproduz exatamente o teste t.

## 3.7 Por que regressão substitui o teste t?

|Por que|Substitui|
|-|-|
|Porque a regressão testa automaticamente:|$\to\qquad H_0: \beta_1 = 0 $|
|Que é equivalente a:|$\to\qquad H_0: \mu_{\text{female}} = \mu_{\text{male}}$|

- E a estatística t é **idêntica**.

#### 🟠 Explicação avançada — Dummy coding e diferença de médias

Se:
- female = 1  
- male = 0  

|Dummy coding|Diferença|
|-|-|
|Então:|$\to\qquad \hat{Y}_{\text{male}} = \beta_0 \\ \to\qquad \hat{Y}_{\text{female}} = \beta_0 + \beta_1$|
|Logo:|$\to\qquad \beta_1 = \hat{Y}_{\text{female}} - \hat{Y}_{\text{male}}$|

- Ou seja, **β₁ é literalmente a diferença de médias**.

Por isso regressão = teste t.

# 4. Regressão no lugar da ANOVA

🔵 **Conteúdo oficial do módulo 6 (IBM)**

O curso explica:

> “Quando comparamos médias de mais de dois grupos, usamos ANOVA.  
> Mas podemos obter exatamente os mesmos resultados usando regressão.”

A ideia é simples:

- ANOVA testa se **todas as médias são iguais**  
- Regressão com variáveis dummy testa **o mesmo**, mas dentro de um modelo linear  

E o mais importante:

> “A estatística F e o valor‑p da regressão são idênticos aos da ANOVA.”

Vamos reproduzir exatamente o exemplo IBM.

## 4.1 Criando grupos de idade (IBM)

O curso divide os instrutores em três grupos:

- **≤ 40 anos**  
- **40–57 anos**  
- **≥ 57 anos**  

Vamos criar essa variável categórica.

```python
bins = [0, 40, 57, 100]
labels = ["<=40", "40-57", ">=57"]

ratings_df["age_group"] = pd.cut(ratings_df["age"], bins=bins, labels=labels, include_lowest=True)
ratings_df[["age", "age_group"]].head()
```

## 4.2 ANOVA tradicional (para comparação)

Vamos rodar ANOVA para `eval` usando `scipy`.

```python
group1 = ratings_df[ratings_df["age_group"] == "<=40"]["eval"]
group2 = ratings_df[ratings_df["age_group"] == "40-57"]["eval"]
group3 = ratings_df[ratings_df["age_group"] == ">=57"]["eval"]

from scipy.stats import f_oneway
f_stat, p_value = f_oneway(group1, group2, group3)
f_stat, p_value
```

🟣 **Interpretação**

- valor‑p > 0.05  
- **não rejeitamos H₀**  
- as médias dos três grupos são estatisticamente iguais  

Agora vamos ver como a regressão chega ao mesmo resultado.

## 4.3 Criando variáveis dummy (IBM)

O curso usa `get_dummies`:

- cada grupo vira uma coluna  
- 1 = pertence ao grupo  
- 0 = não pertence  

Vamos fazer isso.

```python
dummies = pd.get_dummies(ratings_df["age_group"], drop_first=True)
dummies.head()
```

## 4.4 Regressão linear com variáveis dummy

|Modelo:|$$\to\qquad \text{eval} = \beta_0 + \beta_1 D_{40-57} + \beta_2 D_{\ge 57} + \varepsilon$$|
|-|-|

- **Onde:**
    - $\beta_0$ = média do grupo de referência $(≤40)$  
    - $\beta_1$ = diferença entre $40–57 e ≤40$  
    - $\beta_2$ = diferença entre $≥57 e ≤40$  

Vamos ajustar o modelo.

```python
df_reg = pd.concat([ratings_df["eval"], dummies], axis=1)
model_anova_reg = smf.ols("eval ~ Q('40-57') + Q('>=57')", data=df_reg).fit()
model_anova_reg.summary()
```

🟣 **Interpretação**

- Nenhum coeficiente é significativo (valor‑p > 0.05)  
- Isso indica que **nenhum grupo difere do grupo de referência**  
- Conclusão idêntica à ANOVA tradicional  

## 4.5 Obtendo a tabela ANOVA a partir da regressão

O curso IBM mostra como usar `anova_lm` para extrair a tabela ANOVA do modelo OLS.

```python
anova_lm(model_anova_reg)
```

🟣 **Interpretação**

- A estatística F é **idêntica** à da ANOVA tradicional  
- O valor‑p também é **idêntico**  

Isso confirma o que o curso IBM afirma:

> “Regressão e ANOVA produzem exatamente os mesmos resultados.”

## 4.6 Visualização das médias por grupo

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="age_group", y="eval", data=ratings_df, errorbar=('ci', 95))
plt.title("Média de eval por grupo de idade (com IC 95%)")
plt.grid(alpha=0.3)
plt.show()
```

🟣 **Interpretação**

- As médias são visualmente próximas  
- Isso reforça o resultado da ANOVA e da regressão  

## 4.7 Por que regressão substitui ANOVA?

|Por que|Substitui|
|-|-|
|Porque ANOVA testa:|$\to\qquad H_0: \mu_1 = \mu_2 = \mu_3$|
|E regressão testa:|$\to\qquad H_0: \beta_1 = \beta_2 = 0$|
|Mas:|$\to\qquad \beta_1 = \mu_2 - \mu_1 \\ \to\qquad \beta_2 = \mu_3 - \mu_1$|

- Logo:
> - “Testar β = 0 é o mesmo que testar igualdade de médias.”

#### 🟠 Explicação avançada — ANOVA como regressão

- A ANOVA clássica é um caso especial de regressão linear onde:  
    - X são variáveis dummy  
    - Y é contínua  

|ANOVA|Regressão|
|-|-|
|A estatística F da ANOVA é:|$$\to\qquad F = \frac{\text{variância explicada}}{\text{variância residual}}$$|
|A estatística F da regressão é:|$$\to\qquad F = \frac{MS_{\text{modelo}}}{MS_{\text{erro}}}$$|

- As duas expressões são **idênticas**.

**Por isso:**

> “Toda ANOVA é uma regressão.  
> Nem toda regressão é uma ANOVA.”

# 5. Regressão no lugar da Correlação

🔵 **Conteúdo oficial do módulo 6 (IBM)**

O curso explica:

> “Podemos usar regressão linear no lugar do teste de correlação de Pearson.”

A ideia é simples:

- Pearson mede **força e direção** da relação linear  
- Regressão mede **força, direção e magnitude do efeito**  
- O valor‑p da regressão é **idêntico** ao valor‑p do teste de Pearson  
- O $R^2$ da regressão é **$r^2$**  

Vamos reproduzir exatamente o exemplo IBM:

**Pergunta:**  
Existe relação linear entre `beauty` e `eval`?

## 5.1 Correlação de Pearson (método tradicional)

Primeiro, vamos calcular o coeficiente de Pearson.

```python
beauty = ratings_df["beauty"]
evals  = ratings_df["eval"]

pearson_r, p_value = pearsonr(beauty, evals)
pearson_r, p_value
```

🟣 **Interpretação**

- **r ≈ 0.189** → correlação fraca positiva  
- **valor‑p < 0.05** → estatisticamente significativa  

Agora vamos ver como a regressão chega ao mesmo resultado.

## 5.2 Regressão linear simples

|Modelo:|$$\to\qquad \text{eval} = \beta_0 + \beta_1 \cdot \text{beauty} + \varepsilon$$|
|-|-|

```python
model_corr = smf.ols("eval ~ beauty", data=ratings_df).fit()
model_corr.summary()
```

## 5.3 Comparando regressão e Pearson

🔵 **Conteúdo IBM**

O curso afirma:

> “O valor‑p da regressão é o mesmo do teste de Pearson.”

Vamos verificar:

```python
model_corr.pvalues["beauty"]
```

🟣 **Resultado**

✔️ **Os valores‑p são idênticos**  
✔️ **A conclusão estatística é a mesma**  

Isso confirma o que o curso IBM demonstra.

## 5.4 Relação entre R² e r

O curso IBM mostra:

> “Se tirarmos a raiz quadrada de R², obtemos o coeficiente de Pearson.”

Vamos verificar:

```python
r_from_r2 = np.sqrt(model_corr.rsquared)
pearson_r, r_from_r2
```

🟣 **Interpretação**

✔️ **$r = \sqrt{R^2}$**  
✔️ Isso confirma a equivalência matemática entre regressão e correlação linear.

## 5.6 Por que regressão substitui o teste de correlação?

|Por que|Substitui|
|-|-|
|Porque a regressão testa automaticamente:|$\to\qquad H_0: \beta_1 = 0 $|
|E o teste de Pearson testa:|$\to\qquad H_0: \rho = 0 $|
|Mas:|$$\to\qquad \beta_1 = r \cdot \frac{s_y}{s_x}$$|

- Logo:
    - $\beta_1 ≠ 0 ⇔ r ≠ 0$  
    - valor‑p da regressão $=$ valor‑p de Pearson  
    - $R^2 = r^2$  

#### 🟠 Explicação avançada — Conexão matemática entre regressão e Pearson

|Conexão|Matemática|
|-|-|
|A fórmula do coeficiente de regressão é:|$$\to\qquad\qquad \beta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})} {\sum (x_i - \bar{x})^2} $$|
|A fórmula do coeficiente de Pearson é:|$$\to\qquad r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})} {\sqrt{\sum (x_i - \bar{x})^2} \sqrt{\sum (y_i - \bar{y})^2}}$$|
|Portanto:|$\to\qquad \beta_1 = r \cdot \frac{s_y}{s_x} $|
|e:|$$\to\qquad R^2 = r^2 $$|

- Isso mostra que regressão e correlação são **duas faces da mesma moeda**.

# 8. Apêndice Matemático Avançado — Regressão Linear

Este apêndice aprofunda a teoria por trás da regressão linear:

- mínimos quadrados ordinários (OLS)  
- solução matricial  
- distribuição dos coeficientes  
- relação com correlação  
- relação com teste t  
- relação com ANOVA  
- R² e R² ajustado  
- variância dos coeficientes  
- multicolinearidade  

Tudo aqui é opcional, mas extremamente útil para dominar regressão.

## 8.1 O modelo linear

|modelo|linear|
|-|-|
|O modelo de regressão linear simples é:|$\to\qquad Y = \beta_0 + \beta_1 X + \varepsilon$|
|O modelo múltiplo é:|$\to\qquad Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_k X_k + \varepsilon$|
|Em forma matricial:|$\to\qquad \mathbf{Y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}$|

## 8.2 Mínimos Quadrados Ordinários (OLS)

|Mínimo|quadrado|
|-|-|
|O método OLS escolhe $\beta$ que minimiza:|$\to\qquad \sum_{i=1}^n (Y_i - \hat{Y}_i)^2$|
|A solução matricial é:|$\to\qquad \hat{\beta} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{Y}$|

🟠 Demonstração — derivando a solução OLS

|Derivando|OLS|
|-|-|
|1. Escrevemos o erro quadrático:|$\to\qquad S(\beta) = (Y - X\beta)^\top (Y - X\beta)$|
|2. Derivamos em relação a β:|$\to\qquad \frac{\partial S}{\partial \beta} = -2X^\top(Y - X\beta)$|
|3. Igualamos a zero:|$\to\qquad X^\top Y = X^\top X \beta$|
|4. Resolvemos:|$\to\qquad \hat{\beta} = (X^\top X)^{-1} X^\top Y$|



## 8.3 Distribuição dos coeficientes

|Sob hipóteses clássicas:|$\to\qquad \hat{\beta} \sim N(\beta, \sigma^2 (X^\top X)^{-1})$|
|-|-|

- Isso permite:
    - construir intervalos de confiança  
    - realizar testes t  
    - realizar testes F  

## 8.4 Teste t na regressão

|Regressão|Teste t|
|-|-|
|A regressão testa automaticamente:|$\to\qquad H_0: \beta_j = 0$|
|A estatística t é:|$$\to\qquad t = \frac{\hat{\beta}_j}{\text{SE}(\hat{\beta}_j)}$$|

- Isso é exatamente o mesmo teste t do módulo 5.

## 8.5 Teste F na regressão

|Regressão|Teste F|
|-|-|
|A regressão também testa:|$\to\qquad H_0: \beta_1 = \beta_2 = \dots = \beta_k = 0$|
|A estatística F é:|$$\to\qquad F = \frac{MS_{\text{modelo}}}{MS_{\text{erro}}}$$|

- Isso é exatamente o mesmo F da ANOVA.

## 8.6 Relação entre regressão e correlação

|Regressão|Correlação|
|-|-|
|Para regressão simples:|$$\to\qquad \beta_1 = r \cdot \frac{s_y}{s_x}$$|
|E:|$\to\qquad R^2 = r^2$|

#### 🟠 Demonstração — por que R² = r²?

|Por que|$R^2$|
|-|-|
|Em regressão simples:|$$\to\qquad R^2 = \frac{\text{variância explicada}}{\text{variância total}}$$|
|e:|$$\to\qquad  r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})} {\sqrt{\sum (x_i - \bar{x})^2} \sqrt{\sum (y_i - \bar{y})^2}}$$|
|Substituindo β₁ na fórmula de R², obtemos:|$$\to\qquad R^2 = r^2$$|



## 8.7 R² e R² ajustado

- **R²** mede proporção da variância explicada.  
- **R² ajustado** penaliza modelos com muitas variáveis.

|Fórmula:|$$\to\qquad R^2_{\text{adj}} = 1 - (1 - R^2)\frac{n - 1}{n - k - 1}$$|
|-|-|

## 8.8 Variância dos coeficientes

|A variância de $\hat{\beta}$ é:|$\to\qquad \text{Var}(\hat{\beta}) = \sigma^2 (X^\top X)^{-1}$|
|-|-|

- Isso explica:
    - por que multicolinearidade aumenta SE  
    - por que coeficientes deixam de ser significativos  

## 8.9 Multicolinearidade

Ocorre quando variáveis explicativas são altamente correlacionadas.

- Consequências:
    - coeficientes instáveis  
    - erros‑padrão inflados  
    - testes t deixam de ser significativos  
    - R² continua alto  

- Indicadores:
    - VIF (Variance Inflation Factor)  
    - correlação alta entre colunas de X  

## 8.10 Interpretação geométrica da regressão

- A regressão projeta Y no subespaço gerado pelas colunas de X.
    - $\hat{Y}$ é a projeção ortogonal  
    - os resíduos são ortogonais ao espaço de X  

|Formalmente:|$\to\qquad X^\top (Y - \hat{Y}) = 0$|
|-|-|

#### 🟠 Demonstração — resíduos são ortogonais a X

|Residuos|Ortogonais|
|-|-|
|Da equação normal:|$\to\qquad X^\top(Y - X\hat{\beta}) = 0$|
|Mas:|$\to\qquad X\hat{\beta} = \hat{Y}$|
|Logo:|$\to\qquad X^\top(Y - \hat{Y}) = 0$|

Isso significa que os resíduos são perpendiculares ao espaço das variáveis explicativas.


## 8.11 Decomposição da variância

|A regressão decompõe a variância total:|$$\to\qquad SST = SSR + SSE$$|
|-|-|

- Onde:
    - SST = soma total dos quadrados  
    - SSR = soma dos quadrados explicados  
    - SSE = soma dos quadrados do erro  

## 8.12 Resumo visual — Normal, t e F

```python
import numpy as np
from scipy.stats import norm, t, f

x = np.linspace(-4, 4, 400)

plt.figure(figsize=(7, 3))
plt.plot(x, norm.pdf(x), label="Normal", linewidth=2)
plt.plot(x, t.pdf(x, df=5), label="t (df=5)")
plt.plot(x, t.pdf(x, df=30), label="t (df=30)")
plt.title("Comparação: Normal vs t-distribution")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Regressão 💡

A ferramenta que unifica z‑score, Teste A/B e ANOVA**  
A regressão é chamada no curso da IBM de:  
👉 “O carro‑chefe da análise estatística.”

E não é exagero.  
Ela integra os três conceitos que vimos nos últimos posts:

1. z‑score → distância da média  
▪️ A regressão: mede quanto Y muda → quando X muda.  
Isso é, essencialmente, medir distância em unidades padronizadas.

2. Teste A/B → diferença entre dois grupos  
▪️ Quando transformamos um grupo em variável dummy (0|1),  
a regressão faz exatamente o mesmo cálculo do teste t.

→ O valor‑p é idêntico  
→ A estatística t é idêntica  
→ A conclusão é idêntica

3. ANOVA → diferença entre vários grupos  
▪️ Quando usamos várias dummies,  
a regressão reproduz o ANOVA linha por linha.

→ O F‑score é o mesmo  
→ O p‑value é o mesmo  
→ A interpretação é a mesma

🔵 O que muda então?  
▪️ A regressão faz tudo isso ao mesmo tempo:  
compara grupos | mede efeitos | estima relações | testa hipóteses | prevê valores | controla variáveis | quantifica incerteza  
👉 É por isso que ela é o motor da estatística moderna.

📈  Regressão não é só mais um método.  
Ela é o ponto de convergência entre:  
probabilidade | testes de hipótese | comparação de grupos | análise de variância | correlação | previsão  
É a ferramenta que transforma estatística em decisão.

## Regressão 2  💡

📉 Como interpretar uma regressão na prática  
A regressão linear simples responde a uma pergunta direta:  
❓ “Quanto Y muda quando X muda?”

No exemplo do curso IBM o modelo encontrou:  
▪️ Intercepto: 3.998  
▪️ Coeficiente: 0.133  
▪️ p‑value: < 0.001  
▪️ R²: 0.036

🔍 O que isso significa?  
1. O coeficiente (β₁ = 0.133)  
→ Cada 1 ponto de beleza aumenta a avaliação em ~0.13 pontos.  
→ A relação existe, mas é fraca.

2. O p‑value (< 0.001)  
→ A relação é estatisticamente significativa.  
→ Não é fruto do acaso.

3. O R² (0.036)  
→ A beleza explica apenas 3,6% da variação das avaliações.  
→ Ou seja: existe efeito, mas é pequeno.

Esse modelo é perfeita para o post porque mostra:  
coeficientes | erros padrão | t‑stat | p‑values | R² | F‑statistic  
É exatamente o que um analista olha no dia a dia.

📈 A regressão  
→ mostra se existe relação,  
→ quanto ela vale,  
→ quão confiável ela é,  
→ e quanto do fenômeno ela explica.  
👉 É a ferramenta que transforma dados em entendimento real.
