# 2. Intuição das Redes Neurais

Antes de entrar em fórmulas, vamos entender a ideia central.

Redes neurais são modelos que:

- combinam entradas com pesos  
- aplicam uma função de ativação  
- aprendem padrões complexos  
- ajustam pesos automaticamente via gradiente descendente  

A ideia é inspirada (de forma MUITO simplificada) no cérebro humano:
- neurônios recebem sinais  
- somam esses sinais  
- disparam um output  

Redes neurais fazem exatamente isso — mas com matemática.

---
# 2.1 O que é uma rede neural?

Uma rede neural é composta por:

- **camada de entrada**  
- **camadas ocultas**  
- **camada de saída**  

Cada camada tem neurônios conectados entre si.

Visualmente:

```
Entrada → [Camada Oculta 1] → [Camada Oculta 2] → Saída
```

Cada conexão tem um **peso**.

O objetivo do treinamento é:

> Encontrar os pesos que minimizam o erro do modelo.

---
# 2.2 Redes neurais como regressão logística “turbinada”

A regressão logística faz:

```
y = sigmoid(w1*x1 + w2*x2 + ... + b)
```

Uma rede neural faz:

```
h1 = ReLU(w1*x + b1)
h2 = ReLU(w2*h1 + b2)
y  = sigmoid(w3*h2 + b3)
```

Ou seja:

✔️ várias camadas  
✔️ várias transformações  
✔️ funções de ativação não lineares  

Isso permite aprender padrões muito mais complexos.

---
# 2.3 Por que redes neurais são tão poderosas?

Porque elas conseguem aprender:

- não linearidades  
- interações entre variáveis  
- padrões complexos  
- fronteiras de decisão sofisticadas  

E tudo isso **automaticamente**, sem engenharia manual.

---
# 2.4 Redes neurais como composição de funções

Uma rede neural é:

$$
f(x) = f_3(f_2(f_1(x)))
$$

Cada camada aplica uma transformação.

Isso é chamado de **modelo profundo** (deep learning).

---
# 2.5 O que uma rede neural aprende?

Ela aprende:

- pesos (importância das entradas)  
- vieses (ajustes finos)  
- representações internas (features aprendidas)  

Diferente de modelos lineares, ela **cria suas próprias features**.

---
# 2.6 Como redes neurais aprendem?

O processo é:

1. Começa com pesos aleatórios  
2. Faz uma previsão  
3. Calcula o erro  
4. Propaga o erro para trás (backpropagation)  
5. Ajusta os pesos usando gradiente descendente  
6. Repete milhares de vezes  

No final, os pesos convergem para valores que minimizam o erro.

---
# 2.7 Redes neurais vs Boosting

| Característica | Boosting | Redes Neurais |
|----------------|----------|----------------|
| Base | Árvores | Camadas de neurônios |
| Não linearidade | Sim | Sim |
| Interações | Automáticas | Automáticas |
| Escalabilidade | Excelente | Excelente |
| Dados tabulares | Geralmente melhor | Pode ser pior |
| Imagens/Texto | Fraco | Excelente |

Redes neurais brilham em:
- imagens  
- áudio  
- texto  
- séries temporais complexas  

Boosting brilha em:
- dados tabulares estruturados  

---
# 2.8 Visualização simples de uma rede neural

````
import numpy as np

# Exemplo simples de rede neural com pesos fixos
x = np.array([0.5, 1.2])  # duas entradas

W1 = np.array([[0.3, -0.2],
               [0.8,  0.5]])  # pesos da camada oculta

b1 = np.array([0.1, -0.1])

h = np.maximum(0, W1 @ x + b1)  # ReLU

W2 = np.array([0.4, -0.6])
b2 = 0.2

y = 1 / (1 + np.exp(-(W2 @ h + b2)))  # sigmoid

y
````

🟣 **Interpretação**

Mesmo com pesos aleatórios:
- a rede transforma as entradas  
- aplica não linearidades  
- gera uma previsão  

Isso é exatamente o que Keras fará automaticamente.

---
# 2.9 Conclusão desta parte

✔️ Aprendem pesos automaticamente  
✔️ Capturam padrões complexos  
✔️ São generalizações da regressão logística  
✔️ São baseadas em gradiente descendente

# 3. O Neurônio Artificial (Perceptron)

O Perceptron é o bloco fundamental das redes neurais.
Ele foi criado em 1957 por Frank Rosenblatt e é a base de:

- redes neurais profundas  
- MLPs  
- CNNs  
- RNNs  
- Transformers  

Tudo começa aqui.

---
# 3.1 O que é um neurônio artificial?

Um neurônio artificial faz três coisas:

1. **Recebe entradas**  
2. **Multiplica por pesos e soma**  
3. **Aplica uma função de ativação**  

Matematicamente:

$$
z = w_1 x_1 + w_2 x_2 + \dots + w_n x_n + b
$$

$$
y = \sigma(z)
$$

Onde:
- \( w_i \) = pesos  
- \( b \) = viés  
- \( \sigma \) = função de ativação  

---
# 3.2 O Perceptron como classificador linear

O Perceptron original usava:

- soma ponderada  
- função de ativação degrau (step function)  

Ele só conseguia resolver problemas **linearmente separáveis**.

Exemplo:
- AND → funciona  
- OR → funciona  
- XOR → NÃO funciona  

Isso levou ao famoso “inverno das redes neurais” nos anos 70.

---
# 3.3 O Perceptron moderno (neurônio atual)

Hoje usamos funções de ativação suaves:

- sigmoid  
- tanh  
- ReLU  
- GELU  

Isso permite:
- gradiente descendente  
- backpropagation  
- aprendizado profundo  

O neurônio moderno é muito mais poderoso que o Perceptron original.

---
# 3.4 Visualização matemática de um neurônio

````
import numpy as np

# Entradas
x = np.array([0.7, -1.2, 0.3])

# Pesos
w = np.array([0.5, -0.8, 0.1])

# Viés
b = 0.2

# Soma ponderada
z = np.dot(w, x) + b

# Função de ativação (sigmoid)
y = 1 / (1 + np.exp(-z))

z, y
````

🟣 **Interpretação**

- `z` é a combinação linear  
- `y` é a saída ativada  

Isso é exatamente o que cada neurônio faz dentro de uma rede neural.

---
# 3.5 O neurônio como transformador de espaço

O neurônio pega um vetor de entrada e o transforma em:

- outra escala  
- outra direção  
- outra região do espaço  

Com várias camadas, a rede neural cria **representações internas**.

Isso é o coração do deep learning.

---
# 3.6 O Perceptron como fronteira de decisão

O Perceptron define uma fronteira linear:

$$
w_1 x_1 + w_2 x_2 + b = 0
$$

Isso é um hiperplano que separa as classes.

Em 2D → uma linha  
Em 3D → um plano  
Em nD → um hiperplano  

---
# 3.7 Exemplo visual simples (2D)

````
import matplotlib.pyplot as plt

# Pontos de exemplo
X_plot = np.array([
    [1, 2], [2, 3], [3, 3],
    [4, 1], [5, -1], [6, -2]
])
y_plot = np.array([1, 1, 1, 0, 0, 0])

# Pesos arbitrários
w = np.array([1, -1])
b = -1

# Fronteira de decisão: w1*x1 + w2*x2 + b = 0
x1 = np.linspace(0, 7, 100)
x2 = -(w[0] * x1 + b) / w[1]

plt.figure(figsize=(6, 4))
plt.scatter(X_plot[:, 0], X_plot[:, 1], c=y_plot, cmap="coolwarm", s=80)
plt.plot(x1, x2, color="black", label="Fronteira de decisão")
plt.title("Perceptron como classificador linear")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

🟣 **Interpretação**

O Perceptron tenta encontrar uma linha que separa as classes.

Se as classes forem separáveis → funciona.  
Se não forem → falha (ex: XOR).  

---
# 3.8 O problema do XOR (e o nascimento do Deep Learning)

XOR é um problema simples, mas **não linearmente separável**.

O Perceptron falha.

Mas…

Se adicionarmos **uma camada oculta**, o problema é resolvido.

Isso levou ao Perceptron Multicamadas (MLP), que veremos na próxima parte.

---
# 3.9 Conclusão desta parte

✔️ O Perceptron é o neurônio artificial  
✔️ Ele combina entradas com pesos  
✔️ Aplica uma função de ativação  
✔️ É um classificador linear  
✔️ Não resolve problemas não lineares sozinho  
✔️ Mas é o bloco fundamental das redes neurais modernas

# 4. Funções de Ativação

Funções de ativação são o coração das redes neurais.

Elas introduzem **não linearidade**, permitindo que a rede:
- aprenda padrões complexos  
- represente fronteiras de decisão sofisticadas  
- combine features de forma não linear  

Sem ativação → a rede vira uma regressão linear gigante.

---
# 4.1 Sigmoid

A sigmoid foi a primeira função de ativação usada em redes neurais.

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Características:
- saída entre 0 e 1  
- boa para probabilidades  
- usada na **camada de saída** em classificação binária  

Problemas:
- gradiente desaparece  
- saturação em valores extremos  
- lenta para treinar redes profundas

````
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6, 400)
sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(6, 4))
plt.plot(x, sigmoid)
plt.title("Função Sigmoid")
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.2 Tanh

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

Características:
- saída entre -1 e 1  
- centrada em zero (melhor que sigmoid)  

Problemas:
- ainda sofre com gradiente desaparecendo  

Hoje é usada mais em RNNs clássicas.

````
tanh = np.tanh(x)

plt.figure(figsize=(6, 4))
plt.plot(x, tanh, color="orange")
plt.title("Função Tanh")
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.3 ReLU (a revolução do Deep Learning)

A função que mudou tudo:

$$
\text{ReLU}(x) = \max(0, x)
$$

Características:
- simples  
- rápida  
- não satura para valores positivos  
- gradiente não desaparece facilmente  

Problemas:
- neurônios podem “morrer” (ficar sempre em zero)  

ReLU é a ativação padrão em redes profundas.

````
relu = np.maximum(0, x)

plt.figure(figsize=(6, 4))
plt.plot(x, relu, color="green")
plt.title("Função ReLU")
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.4 Leaky ReLU

Resolve o problema do “neurônio morto”:

$$
\text{LeakyReLU}(x) = 
\begin{cases}
x, & x > 0 \\
0.01x, & x \le 0
\end{cases}
$$

````
leaky_relu = np.where(x > 0, x, 0.01 * x)

plt.figure(figsize=(6, 4))
plt.plot(x, leaky_relu, color="purple")
plt.title("Função Leaky ReLU")
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.5 GELU (ativação usada no Transformer)

GELU = Gaussian Error Linear Unit

É a ativação usada no BERT, GPT, Transformers modernos.

Aproximação:

$$
\text{GELU}(x) \approx 0.5x(1 + \tanh(\sqrt{2/\pi}(x + 0.0447x^3)))
$$

Características:
- suave  
- probabilística  
- excelente para redes profundas  

É a ativação padrão em modelos de linguagem.

````
gelu = 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.0447 * x**3)))

plt.figure(figsize=(6, 4))
plt.plot(x, gelu, color="red")
plt.title("Função GELU")
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.6 Comparação visual

````
plt.figure(figsize=(8, 5))
plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="Tanh")
plt.plot(x, relu, label="ReLU")
plt.plot(x, leaky_relu, label="Leaky ReLU")
plt.plot(x, gelu, label="GELU")
plt.title("Comparação de Funções de Ativação")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

---
# 4.7 Quando usar cada ativação?

| Ativação | Onde usar |
|----------|-----------|
| **ReLU** | camadas ocultas (padrão) |
| **Leaky ReLU** | quando ReLU “mata” neurônios |
| **Sigmoid** | saída de classificação binária |
| **Tanh** | RNNs clássicas |
| **GELU** | Transformers, redes profundas modernas |

Em 90% dos casos:
- ReLU nas camadas ocultas  
- Sigmoid na saída (classificação binária)  
- Softmax na saída (classificação multiclasse)  

---
# 4.8 Conclusão desta parte

✔️ Funções de ativação dão não linearidade às redes  
✔️ ReLU é a ativação padrão moderna  
✔️ Sigmoid é usada na saída binária  
✔️ GELU é usada em Transformers  
✔️ Ativação correta = rede mais estável e eficiente

# 5. Perceptron Multicamadas (MLP) e Backpropagation

O Perceptron Multicamadas (MLP) é a primeira rede neural “de verdade”.

Ele resolve o maior problema do Perceptron simples:

> **não consegue aprender padrões não lineares (ex: XOR)**.

Com uma camada oculta, tudo muda.

---
# 5.1 Estrutura de um MLP

Um MLP típico tem:

- camada de entrada  
- **uma ou mais camadas ocultas**  
- camada de saída  

Exemplo:

```
Entrada → [Oculta 1] → [Oculta 2] → Saída
```

Cada camada:
- aplica pesos  
- soma  
- ativa  

Isso cria **representações internas**.

---
# 5.2 Por que o MLP resolve o XOR?

Porque a camada oculta cria **features intermediárias**.

A rede aprende:
- combinações não lineares  
- interações entre variáveis  
- fronteiras complexas  

Isso é impossível para um Perceptron simples.

---
# 5.3 Matemática do MLP

Para uma camada oculta:

$$
h = \sigma(W_1 x + b_1)
$$

Para a saída:

$$
y = \sigma(W_2 h + b_2)
$$

Onde:
- \( W_1, W_2 \) são matrizes de pesos  
- \( b_1, b_2 \) são vieses  
- \( \sigma \) é a ativação (ReLU, sigmoid etc.)  

---
# 5.4 Backpropagation — como a rede aprende?

Backpropagation é o algoritmo que ajusta os pesos.

O processo:

1. **Forward pass**  
   - calcula a previsão  

2. **Calcula o erro**  
   - diferença entre previsão e rótulo  

3. **Backward pass**  
   - propaga o erro para trás  
   - calcula gradientes  

4. **Atualiza os pesos**  
   - usando gradiente descendente  

Isso se repete milhares de vezes.

---
# 5.5 Exemplo numérico simples de MLP (1 camada oculta)

````
# Entradas
x = np.array([0.5, -1.0])

# Pesos da camada oculta (2 neurônios)
W1 = np.array([
    [0.3, -0.2],
    [0.7,  0.4]
])
b1 = np.array([0.1, -0.1])

# Forward da camada oculta (ReLU)
h = np.maximum(0, W1 @ x + b1)

# Pesos da camada de saída
W2 = np.array([0.6, -0.5])
b2 = 0.2

# Saída (sigmoid)
y = 1 / (1 + np.exp(-(W2 @ h + b2)))

h, y
````

🟣 **Interpretação**

- `h` são as features aprendidas pela camada oculta  
- `y` é a previsão final  

A rede está criando **representações internas** — isso é Deep Learning.

---
# 5.6 Visualização: MLP aprendendo uma fronteira não linear

````
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons

# Dataset não linear
X_moons, y_moons = make_moons(n_samples=400, noise=0.2, random_state=42)

# Treinando um MLP
mlp = MLPClassifier(
    hidden_layer_sizes=(10, 10),
    activation="relu",
    max_iter=2000,
    random_state=42
)
mlp.fit(X_moons, y_moons)

# Grade para visualização
xx, yy = np.meshgrid(
    np.linspace(X_moons[:,0].min()-1, X_moons[:,0].max()+1, 300),
    np.linspace(X_moons[:,1].min()-1, X_moons[:,1].max()+1, 300)
)
grid = np.c_[xx.ravel(), yy.ravel()]
Z = mlp.predict(grid).reshape(xx.shape)

plt.figure(figsize=(7, 5))
plt.contourf(xx, yy, Z, cmap="coolwarm", alpha=0.4)
plt.scatter(X_moons[:,0], X_moons[:,1], c=y_moons, cmap="coolwarm", s=40)
plt.title("MLP aprendendo uma fronteira não linear")
plt.show()
````

🟣 **Interpretação**

O MLP aprende uma fronteira de decisão **curvada**, impossível para modelos lineares.

Isso é o poder da não linearidade + camadas ocultas.

---
# 5.7 Por que o MLP é tão importante?

Porque ele é a base de:

- redes profundas (DNNs)  
- redes convolucionais (CNNs)  
- redes recorrentes (RNNs)  
- Transformers  

Todos esses modelos são apenas **variações do MLP**.

---
# 5.8 Conclusão desta parte

✔️ O MLP resolve problemas não lineares  
✔️ Ele cria representações internas  
✔️ Backpropagation ajusta os pesos  
✔️ MLP é a base de todo Deep Learning moderno

# 6. TensorFlow/Keras — Construindo sua primeira rede neural

Agora vamos treinar nossa primeira rede neural real usando:

- TensorFlow  
- Keras (API de alto nível)  

Keras facilita tudo:
- define camadas  
- compila o modelo  
- treina com backpropagation  
- calcula métricas  

Vamos usar o mesmo dataset `ratings_df` para comparar com boosting.

---
# 6.1 Importando TensorFlow/Keras

````
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
````

---
# 6.2 Preparando os dados

Redes neurais **precisam de dados escalados**.
Vamos usar `StandardScaler`.

````
X = ratings_df[["beauty", "age", "students", "female_dummy"]]
y = ratings_df["high_eval"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled[:5]
````

---
# 6.3 Construindo o modelo Keras

Vamos criar um MLP simples:

- 1ª camada oculta: 16 neurônios, ReLU  
- 2ª camada oculta: 8 neurônios, ReLU  
- Saída: 1 neurônio, Sigmoid  

Esse é um modelo clássico para classificação binária.

````
model = keras.Sequential([
    layers.Dense(16, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    layers.Dense(8, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

model.summary()
````

---
# 6.4 Compilando o modelo

Precisamos definir:

- **loss** → função de perda  
- **optimizer** → algoritmo de atualização  
- **metrics** → métricas monitoradas  

Para classificação binária:
- perda: `binary_crossentropy`  
- otimizador: `adam` (padrão moderno)  

````
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
````

---
# 6.5 Treinando o modelo

Vamos treinar por 50 épocas.

Uma época = passar por todo o dataset uma vez.

````
history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=0
)
````

---
# 6.6 Curvas de treinamento

````
plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Treino")
plt.plot(history.history["val_accuracy"], label="Validação")
plt.title("Acurácia — Treino vs Validação")
plt.xlabel("Épocas")
plt.ylabel("Acurácia")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

# 8. Ajuste de Hiperparâmetros em Redes Neurais

Redes neurais têm muitos hiperparâmetros importantes:

- número de camadas  
- número de neurônios  
- função de ativação  
- taxa de aprendizado (learning rate)  
- tamanho do batch  
- regularização (L2, Dropout)  
- número de épocas  

Nesta parte, vamos explorar os mais importantes na prática.

---
# 8.1 Dropout — a arma secreta contra overfitting

Dropout desliga aleatoriamente neurônios durante o treino.

Exemplo:
- taxa = 0.3 → 30% dos neurônios desligados  

Isso força a rede a:
- não memorizar o treino  
- aprender representações mais robustas  

É uma das técnicas mais importantes do deep learning moderno.

````
model_dropout = keras.Sequential([
    layers.Dense(32, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(1, activation="sigmoid")
])

model_dropout.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

history_dropout = model_dropout.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=80,
    batch_size=16,
    verbose=0
)
````

---
# 8.2 Curvas de aprendizado com Dropout

````
plt.figure(figsize=(8, 5))
plt.plot(history_dropout.history["accuracy"], label="Treino")
plt.plot(history_dropout.history["val_accuracy"], label="Validação")
plt.title("Acurácia — com Dropout")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history_dropout.history["loss"], label="Treino")
plt.plot(history_dropout.history["val_loss"], label="Validação")
plt.title("Perda — com Dropout")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

🟣 **Interpretação**

- Treino e validação ficam mais próximos  
- Overfitting reduz drasticamente  
- A rede generaliza melhor

---
# 8.4 Batch Size — o hiperparâmetro mais subestimado

O batch size controla quantas amostras são usadas antes de atualizar os pesos.

| Batch | Características |
|-------|-----------------|
| **Pequeno (8–32)** | mais ruído, melhor generalização |
| **Médio (64–128)** | equilíbrio |
| **Grande (256+)** | rápido, mas pode overfitar |

Vamos testar três tamanhos.

````
batch_sizes = [8, 32, 128]
results_batch = []

for bs in batch_sizes:
    model_bs = keras.Sequential([
        layers.Dense(16, activation="relu", input_shape=(X_train_scaled.shape[1],)),
        layers.Dense(8, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    model_bs.compile(
        loss="binary_crossentropy",
        optimizer="adam",
        metrics=["accuracy"]
    )
    model_bs.fit(
        X_train_scaled, y_train,
        validation_split=0.2,
        epochs=40,
        batch_size=bs,
        verbose=0
    )
    pred = (model_bs.predict(X_test_scaled) > 0.5).astype(int)
    results_batch.append([
        bs,
        accuracy_score(y_test, pred),
        f1_score(y_test, pred)
    ])

pd.DataFrame(results_batch, columns=["Batch Size", "Acurácia", "F1"])
````

🟣 **Interpretação**

- batch pequeno → melhor generalização  
- batch grande → mais rápido, mas pode perder qualidade  

---
# 8.5 Learning Rate — o hiperparâmetro mais importante

O learning rate controla o tamanho dos passos do gradiente.

- muito alto → diverge  
- muito baixo → aprende devagar  

Vamos testar três valores.

````
learning_rates = [0.0005, 0.001, 0.01]
results_lr = []

for lr in learning_rates:
    model_lr = keras.Sequential([
        layers.Dense(16, activation="relu", input_shape=(X_train_scaled.shape[1],)),
        layers.Dense(8, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    model_lr.compile(
        loss="binary_crossentropy",
        optimizer=keras.optimizers.Adam(learning_rate=lr),
        metrics=["accuracy"]
    )
    model_lr.fit(
        X_train_scaled, y_train,
        validation_split=0.2,
        epochs=40,
        batch_size=16,
        verbose=0
    )
    pred = (model_lr.predict(X_test_scaled) > 0.5).astype(int)
    results_lr.append([
        lr,
        accuracy_score(y_test, pred),
        f1_score(y_test, pred)
    ])

pd.DataFrame(results_lr, columns=["Learning Rate", "Acurácia", "F1"])
````

🟣 **Interpretação**

- LR muito alto → instável  
- LR muito baixo → lento  
- LR ideal → entre 0.001 e 0.01  

---
# 8.6 Conclusões práticas de tuning

✔️ **Dropout** reduz overfitting  
✔️ **Batch pequeno** melhora generalização  
✔️ **Learning rate** é o hiperparâmetro mais crítico  
✔️ **EarlyStopping** evita treinos longos e overfitting  
✔️ Redes neurais exigem tuning cuidadoso
